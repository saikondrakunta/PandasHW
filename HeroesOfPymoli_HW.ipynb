```python
import pandas as pd
```

```python
file = "HeroesOfPymoli/Resources/purchase_data.csv"
hero_df = pd.read_csv(file)
hero_df.head()
```

```python
#PlayerCount
total_players = len(hero_df["SN"].value_counts())

```

```python
df_players = pd.DataFrame({"Total Players": [total_players]})
df_players.head()
```

```python
#Purchasing Analysis (Total)
#Number of Unique Items #Average Purchase Price #Total Number of Purchases #Total Revenue
unique_items = len(hero_df["Item ID"].value_counts())
```

```python
total_revenue=hero_df['Price'].sum()
```

```python
number_of_purchase=len(hero_df["Purchase ID"].value_counts())
```

```python
avg_purchase = hero_df["Price"].mean()
```

```python
df_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                   "Average Price": [avg_purchase],
                   "Number of Purchases":[number_of_purchase],
                   "Total Revenue": [total_revenue]})
df_analysis["Average Price"]=df_analysis["Average Price"].astype(float).map("${:,.2f}".format)
df_analysis["Total Revenue"]=df_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
df_analysis
```

```python
#Gender Demographics
#Percentage and Count of Male Players #Percentage and Count of Female Players #Percentage and Count of Other / Non-Disclosed
gender_demographic = hero_df.groupby("Gender")["SN"].nunique()
total_players = len(hero_df["SN"].value_counts())
percent_of_players = (gender_demographic/total_players)*100
gender_df = pd.DataFrame({"Total Count": gender_demographic,
                   "Percentage of Players": percent_of_players})
gender_df
```

```python
#Purchasing Analysis (Gender)
purchase_gender_demographic = hero_df.groupby("Gender")["Purchase ID"].nunique()
```

```python
avg_purchase = hero_df.groupby("Gender")["Price"].mean()
```

```python
total_purchase = hero_df.groupby("Gender")["Price"].sum()
```

```python
avg_total_purchase_per_person_by_gender = total_purchase/gender_demographic
```

```python
df_gender_analysis = pd.DataFrame({"Purchase Count": purchase_gender_demographic,
                   "Average Purchase Price": avg_purchase,
                   "Total Purchase Value":total_purchase,
                   "Avg Total Purchase per Person": avg_total_purchase_per_person_by_gender})
df_gender_analysis["Average Purchase Price"]=df_gender_analysis["Average Purchase Price"].astype(float).map("${:,.2f}".format)
df_gender_analysis["Total Purchase Value"]=df_gender_analysis["Total Purchase Value"].astype(float).map("${:,.2f}".format)
df_gender_analysis["Avg Total Purchase per Person"]=df_gender_analysis["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
df_gender_analysis
```

```python
#Age Demographics
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100 ]
age_bin = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]
hero_df["Categorized Players"]= pd.cut(hero_df["Age"], bins, labels=age_bin)
age_grouped = hero_df.groupby("Categorized Players")
count_age = age_grouped["SN"].nunique()
percentage_by_age = (count_age/total_players)*100
age_demographics = pd.DataFrame({"Total Count": count_age,
                                 "Percentage of Players":percentage_by_age })
age_demographics.index.name= None
age_demographics["Percentage of Players"] =age_demographics["Percentage of Players"].astype(float).map("{:,.2f}%".format)
age_demographics
```

```python
#Puchase Analysis (Age)
purchase_age_count = age_grouped["Purchase ID"].count()
avg_purchase=age_grouped["Price"].mean()
total_purchase_age = age_grouped["Price"].sum()
avg_purchase_age = total_purchase_age/count_age
age_analysis_df = pd.DataFrame ({"Purchase Count":purchase_age_count ,
                                 "Average Purchase Price": avg_purchase,
                                 "Total Purchase Value": total_purchase_age,
                                 "Avg Total Purchase per Person":avg_purchase_age })
age_analysis_df["Average Purchase Price"]= age_analysis_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_analysis_df["Total Purchase Value"]= age_analysis_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_analysis_df["Avg Total Purchase per Person"]= age_analysis_df["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)
age_analysis_df
```

```python
#Top Spender
buyer = hero_df.groupby("SN")
count_buyers = buyer["Purchase ID"].count()
avg_buyers = buyer["Price"].mean()
purchase_total_buyers = buyer["Price"].sum()
top_spender_df = pd.DataFrame ({"Purchase Count": count_buyers,
                                "Average Purchase Price":avg_buyers,
                                "Total Purchase Value":purchase_total_buyers})
filtered_top_spender_df = top_spender_df.sort_values (["Total Purchase Value"], ascending =False)
filtered_top_spender_df["Average Purchase Price"] = top_spender_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
filtered_top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
filtered_top_spender_df.head()
```

```python
#Most Popular Item
items = hero_df[["Item ID", "Item Name", "Price"]]
items_grouped = hero_df.groupby(["Item ID", "Item Name"])
purchase_count = items_grouped["Price"].count()
purchase_price = items_grouped["Price"].sum()
price_per_item = purchase_price/purchase_count
popular_df = pd.DataFrame ({"Purchase Count": purchase_count,
                            "Item Price": price_per_item,
                            "Total Purchase Value": purchase_price})
filtered_popular_df = popular_df.sort_values (["Purchase Count"], ascending =False)
filtered_popular_df["Item Price"]= popular_df["Item Price"].astype(float).map("${:,.2f}".format)
filtered_popular_df["Total Purchase Value"]= popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
filtered_popular_df.head()
```

```python
#Most Profitable Item
filtered_popular_df= popular_df.sort_values (["Total Purchase Value"], ascending =False)
filtered_popular_df["Item Price"]= popular_df["Item Price"].astype(float).map("${:,.2f}".format)
filtered_popular_df["Total Purchase Value"]= popular_df["Total Purchase Value"].astype(float).map("${:,.2f}".format)
filtered_popular_df.head()
```
